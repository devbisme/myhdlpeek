# Introduction

New tools have appeared in the past few years to make design and documentation of digital systems
easier than ever:

* Python-based [hardware description languages](https://www.wikiwand.com/en/Hardware_description_language) like [MyHDL](myhdl.org) and [Amaranth](https://amaranth-lang.org/docs/amaranth/latest/) that can both synthesize and simulate a digital system.
* [Jupyter notebooks](https://jupyter.org/) that combine explanatory text and executable code into a single, interactive document.

But one thing that's missing is an easy way to show the results of digital system simulations inside a notebook. That's where `myhdlpeek` comes in.

`myhdlpeek` lets you monitor signals in a MyHDL or Amaranth digital system simulation
and display them as waveforms or as tables in a Jupyter notebook.

`myhdlpeek` implements a `Peeker` object that monitors a signal and records
the time and value when it changes.
Just add multiple Peekers where you want to monitor something (even at sub-levels
of a hierarchical design) and then view the collected data as waveforms or in a table
with a single command.
You can also select which signals are shown, set the beginning and
ending times of the display, find portions of the waveforms where user-specified triggers occur,
and much more.

# Simple Multiplexer Example

This example shows how to use the `myhdlpeek` module to monitor the inputs and output of a simple, two-input multiplexer.

## MyHDL Version

````python
from myhdl import *                 # Import the MyHDL module.
from myhdlpeek.myhdl import Peeker  # Import the MyHDL version of the myhdlpeeker module.


# Define the multiplexer module.
def mux(z, a, b, sel):
    """A simple multiplexer."""

    @always_comb
    def mux_logic():
        if sel == 1:
            z.next = a  # Signal a sent to mux output when sel is high.
        else:
            z.next = b  # Signal b sent to mux output when sel is low.
           
    return mux_logic

# Create some signals to attach to the multiplexer.
a, b, z = [Signal(0) for _ in range(3)]  # Integer signals for the inputs & output.
sel = Signal(bool(0))                    # Binary signal for the selector.

# Create some Peekers to monitor the multiplexer I/Os.
Peeker(a, 'a')         # Add a Peeker to the a input.
Peeker(b, 'b')         # Add a Peeker to the b input.
Peeker(z, 'z')         # Add a peeker to the z output.
Peeker(sel, 'select')  # Add a Peeker to the select input. The Peeker label doesn't have to match the signal name.

# Instantiate the multiplexer.
mux_1 = mux(z, a, b, sel)

# Create a test process to apply random patterns to the multiplexer.
from random import randrange
def test():
    '''Simple testbed generator that applies random inputs to the multiplexer.'''
    for _ in range(8):
        a.next, b.next, sel.next = randrange(8), randrange(8), randrange(2)
        yield delay(1)

# Create a simulator for the multiplexer, test process and the peekers.
sim = Simulation(mux_1, test(), *Peeker.instances())

# Run the simulation.
sim.run()

# Display the signal values as a table.
Peeker.to_html_table()

# Display the complete waveforms captured by all the Peekers.
Peeker.show_waveforms()

# Clear the Peekers for the next simulation.
Peeker.clear()
````

## Amaranth Version

````python
from amaranth import *                 # Import the Amaranth module.
from amaranth.sim import *             # Import the Amaranth simulator.
from myhdlpeek.amaranth import Peeker  # Import the Amaranth version of the myhdlpeeker module.

# Define the multiplexer module.
class Multiplexer(Elaboratable):
    def __init__(self, width):
        self.a   = Signal(width)
        self.b   = Signal(width)
        self.sel = Signal(1)
        self.z   = Signal(width)
 
    def elaborate(self, platform):
        m = Module()
        m.d.comb += self.z.eq(Mux(self.sel, self.a, self.b))
        return m

# Instantiate the multiplexer with a data width of three bits.
mux = Multiplexer(width=3)

# Create some Peekers to monitor the multiplexer I/Os.
Peeker(mux.a, 'a')         # Add a Peeker to the a input.
Peeker(mux.b, 'b')         # Add a Peeker to the b input.
Peeker(mux.z, 'z')         # Add a peeker to the z output.
Peeker(mux.sel, 'select')  # Add a Peeker to the select input. The Peeker label doesn't have to match the signal name.

# Create a simulator object for the multiplexer.
sim = Simulator(mux)

# Create a test process to apply random patterns to the multiplexer.
def test_proc():
    from random import randint
    while True:
        yield mux.a.eq(randint(0,7))
        yield mux.b.eq(randint(0,7))
        yield mux.sel.eq(randint(0,1))
        yield Delay(0.1)  # Amaranth allows floats for delays.

# Add the test process to the simulator.
sim.add_process(test_proc)

# Assign the simulator object to the Peekers. Run the Peekers as if
# they were VCD writers since this is strictly a combinatorial design.
Peeker.assign_simulator(sim, use_vcd_writer=True)

# Run the simulation for eight cycles.
sim.run_until(1)

# Display the signal values as a table.
Peeker.to_html_table()

# Display the complete waveforms captured by all the Peekers.
Peeker.show_waveforms()

# Clear the Peekers for the next simulation.
Peeker.clear()
````

# Hierarchical Examples

Peekers can also be used to monitor signals within sub-levels of a hierarchical design.
Indices will be added to the signal names to distinguish them from one another.

## An Adder in MyHDL

````python
from myhdl import *
from myhdlpeek.myhdl import Peeker

def adder_bit(a, b, c_in, sum_, c_out):
    '''Single bit adder.'''
    @always_comb
    def adder_logic():
        sum_.next = a ^ b ^ c_in
        c_out.next = (a & b) | (a & c_in) | (b & c_in)
    
    # Add some peekers to monitor the inputs and outputs.
    # These will get replicated for each adder bit.
    Peeker(a, 'a')
    Peeker(b, 'b')
    Peeker(c_in, 'c_in')
    Peeker(sum_, 'sum')
    Peeker(c_out, 'c_out')
    
    return adder_logic

def adder(a, b, sum_):
    '''Connect single-bit adders to create a complete adder.'''
    c = [Signal(bool(0)) for _ in range(len(a)+1)] # Carry signals between stages.
    s = [Signal(bool(0)) for _ in range(len(a))]   # Sum bit for each stage.
    stages = []  # Storage for adder bit instances.
    
    # Create the adder bits and connect them together.
    for i in range(len(a)):
        stages.append( adder_bit(a=a(i), b=b(i), sum_=s[i], c_in=c[i], c_out=c[i+1]) )
    
    # Concatenate the sum bits and send them out on the sum_ output.
    @always_comb
    def make_sum():
        sum_.next = ConcatSignal(*reversed(s))
    
    return instances()  # Return all the adder stage instances.

# Create signals for interfacing to the adder.
a, b, sum_ = [Signal(intbv(0,0,8)) for _ in range(3)]

# Instantiate the adder.
add_1 = adder(a=a, b=b, sum_=sum_)

# Create some more peekers to monitor the top-level buses.
Peeker(a, 'a_bus')
Peeker(b, 'b_bus')
Peeker(sum_, 'sum_bus')

# Create a test process that applies random inputs to the adder.
from random import randrange
def adder_test(num_steps):
    for _ in range(num_steps):
        a.next, b.next = randrange(0, a.max), randrange(0, a.max)
        yield delay(1)

# Simulate the adder, test process and peekers.
Simulation(add_1, adder_test(20), *Peeker.instances()).run()

# Show all the signals in the hierarchical design.
Peeker.show_waveforms()

Peeker.clear()
````

## An ALU in Amaranth

````python
from amaranth import *
from amaranth.sim import *
from myhdlpeek.amaranth import Peeker

class Adder(Elaboratable):
    '''Adder module.'''
    def __init__(self, width):
        self.a   = Signal(width)
        self.b   = Signal(width)
        self.o   = Signal(width)
        
        # Peekers for adder I/O.
        Peeker(self.a, "ADDER_A")
        Peeker(self.b, "ADDER_B")
        Peeker(self.o, "SUM")
 
    def elaborate(self, platform):
        m = Module()
        m.d.comb += self.o.eq(self.a + self.b)
        return m

class Subtractor(Elaboratable):
    '''Subtractor module.'''
    def __init__(self, width):
        self.a   = Signal(width)
        self.b   = Signal(width)
        self.o   = Signal(width)

        # Peekers for subtractor I/O.
        Peeker(self.a, "SUB_A")
        Peeker(self.b, "SUB_B")
        Peeker(self.o, "DIFF")
 
    def elaborate(self, platform):
        m = Module()
        m.d.comb += self.o.eq(self.a - self.b)
        return m

class ALU(Elaboratable):
    '''ALU made up of an adder and a subtractor.'''
    def __init__(self, width):
        self.op  = Signal()
        self.a   = Signal(width)
        self.b   = Signal(width)
        self.o   = Signal(width)
        
        # Peekers for ALU I/O.
        Peeker(self.a, "A")
        Peeker(self.b, "B")
        Peeker(self.op, "ADD/SUB")
        Peeker(self.o, "OUTPUT")
 
        self.add = Adder(width)
        self.sub = Subtractor(width)
 
    def elaborate(self, platform):
        m = Module()
        m.submodules.add = self.add
        m.submodules.sub = self.sub
        m.d.comb += [
            self.add.a.eq(self.a),
            self.sub.a.eq(self.a),
            self.add.b.eq(self.b),
            self.sub.b.eq(self.b),
        ]
        with m.If(self.op):
            # Subtract a-b when op==1.
            m.d.comb += self.o.eq(self.sub.o)
        with m.Else():
            # Add a+b when op==0.
            m.d.comb += self.o.eq(self.add.o)
        return m

alu = ALU(width=4)  # Instantiate a four-bit ALU.

sim = Simulator(alu)

# Create test process fo applying random inputs and add/sub operations.
def test_proc():
    from random import randint
    while True:
        yield alu.a.eq(randint(0,15))
        yield alu.b.eq(randint(0,15))
        yield alu.op.eq(randint(0,1))
        yield Delay(1)

# Add the test process to the simulator.
sim.add_process(test_proc)

# Assign the simulator object to the Peekers. Run the Peekers as if
# they were VCD writers since this is strictly a combinatorial design.
Peeker.assign_simulator(sim, use_vcd_writer=True)

# Run the simulation for twenty cycles.
sim.run_until(20)

# Display the complete waveforms captured by all the Peekers.
# An explicit ordering of the Peekers is given in this case.
Peeker.show_waveforms("ADDER_A ADDER_B SUM SUB_A SUB_B DIFF A B ADD/SUB OUTPUT")

Peeker.clear()
````

# Sequential Examples



## A State Machine in MyHDL

````python
from myhdl import *
from myhdlpeek.myhdl import Peeker

states = enum('S0', 'S1', 'S2', 'S3')

def fsm(clk, sel, out):
    """A simple finite state machine."""
    
    s = Signal(states.S0)

    @always_seq(clk.posedge, reset=None)
    def next_state():
        if sel:
            # S0 -> S1 -> S2 -> S3 -> S0 -> ...
            if s == states.S0:
                s.next = states.S1
            elif s == states.S1:
                s.next = states.S2
            elif s == states.S2:
                s.next = states.S3
            else:
                s.next = states.S0
        else:
            # S0 -> S3 -> S2 -> S1 -> S0 -> ...
            if s == states.S0:
                s.next = states.S3
            elif s == states.S1:
                s.next = states.S0
            elif s == states.S2:
                s.next = states.S1
            else:
                s.next = states.S2
                
    @always_comb
    def output():
        out.next = s
           
    return next_state, output

# Create some signals to attach to the FSM.
clk, sel, out = Signal(bool(0)), Signal(bool(0)), Signal(states.S0)

# Create some Peekers to monitor the FSM I/Os.
Peeker(clk, 'clk')         # Add a Peeker to the a input.
Peeker(sel, 'sel')         # Add a Peeker to the b input.
Peeker(out, 'out')         # Add a peeker to the z output.

# Instantiate the FSM.
fsm_1 = fsm(clk, sel, out)

# Create a test process to cause random state changes.
def test():
    from random import randint
    choices = [states.__dict__[k] for k in states._names]
    for _ in range(10):
        sel.next = randint(0,1)
        clk.next = 0
        yield delay(1)
        clk.next = 1
        yield delay(1)

# Simulate the FSM, test process and the peekers.
sim = Simulation(fsm_1, test(), *Peeker.instances()).run()

# Display the complete waveforms captured by all the Peekers. 
Peeker.show_waveforms()

Peeker.clear()
````

## A Counter in Amaranth

````python
from amaranth import *
from amaranth.sim import *
from myhdlpeek.amaranth import Peeker
  
class Counter(Elaboratable):
    '''A simple counter that increments at every clock cycle.'''
    def __init__(self, width):
        self.v = Signal(width)
        
        # Create monitors for the counter value and individual bits.
        Peeker(self.v, 'count')
        for bit in self.v:
            Peeker(bit, self.v.name + '_bit')
 
    def elaborate(self, platform):
        m = Module()
        m.d.sync += self.v.eq(self.v + 1)
        return m
 
counter = Counter(width=4)

sim = Simulator(counter)

# Add a 1 MHz clock to the simulator.
sim.add_clock(1e-6)

# Assign the simulator object to the Peekers. Since this is a sequential
# design driven by a clock, the Peekers can execute as processes instead
# of as VCD writers.
Peeker.assign_simulator(sim)

# Simulate for 20 microseconds.
sim.run_until(20e-6)

Peeker.show_waveforms()

Peeker.clear()
````

## Selecting Display Type

You can display waveforms using either the [`matplotlib`](https://matplotlib.org/) or
[`wavedrom`](https://wavedrom.com/) libraries.
The `Peeker.show_waveforms(...)` function will use whichever library is active.
`matplotlib` is the default, but that can be changed like so:

```py
Peeker.config_defaults(use_wavedrom=True)
```

To change back to `matplotlib`, use:

```py
Peeker.config_defaults(use_wavedrom=False)
```

If you want to explicitly use `matplotlib` or `wavedrom` for displaying waveforms, that can be done by replacing `Peeker.show_waveforms(...)` with `Peeker.to_matplotlib(...)` or `Peeker.to_wavedrom(...)`, respectively.

If you choose to use `wavedrom`, you also need to set a flag depending upon whether you are using Jupyter or JupyterLab for your notebook:

```py
Peeker.config_defaults(use_wavedrom=True, use_jupyter=True)   # If using Jupyter.

Peeker.config_defaults(use_wavedrom=True, use_jupyter=False)  # If using JupyterLab.
```

## Selecting Waveforms to Display

By default, `show_waveforms()` displays all the captured waveforms. But you can also specify a subset of the waveforms for display:

````python
Peeker.show_waveforms('a_bus', 'b_bus', 'sum_bus', 'sum[2]', 'sum[1]', 'sum[0]');
````

If you don't like typing all those quotation marks, you can place multiple, space-separated Peeker names inside a string:

````python
Peeker.show_waveforms('a_bus b_bus sum_bus sum[2] sum[1] sum[0]');
````

## Spacing the Waveforms

If you want to place some space between the waveforms, just insert a string that doesn't match any peeker's name (I like to use a `|` character for that):

````python
Peeker.show_waveforms('a_bus b_bus | sum_bus sum[2] sum[1] sum[0]');
````

## Specifying a Time Window

To show only a segment of the waveforms, use the `start_time` and `stop_time` options:

````python
signals = 'a_bus b_bus | sum_bus sum[0]'
Peeker.show_waveforms(signals, start_time=5, stop_time=15);
````

## Showing Grid and Cycle Times

If it's hard to tell the interval covered by the waveforms, you can turn on the display of cycle times using the `tick` or `tock` options:

````python
Peeker.show_waveforms(signals, start_time=5, stop_time=15, tick=True);
````

````python
Peeker.show_waveforms(signals, start_time=5, stop_time=15, tock=True);
````

You can set the format for the `tick`/`tock` labels using the optional `time_fmt` argument ([here](https://matplotlib.org/3.2.1/api/text_api.html#matplotlib.text.Text) is a list of all the formatting options):

````python
Peeker.show_waveforms(signals, stop_time=15, tock=True,
                      time_fmt={'color':'#080', 'fontweight':'heavy', 'fontstyle':'oblique'});
````

# Triggers

If you capture large signal traces, it's useful to be able to find portions where particular things happen.
*Triggers* are used for this.
These are analogous to the trigger function on logic analyzers and oscilloscopes.

Triggers are created by performing operations on `Trace` objects.
A `Trace` is just a list of values for a signal and the times at which they occurred.
Every `Peeker` object contains a `Trace` to record the values observed by the `Peeker`.

Arithmetic (`+, -, *, %, /, //, **, <<, >>, abs`), logical (`&, |, ^, ~`), and comparison (`==, !=, <, >, <=, >=`) operations
can be performed on `Trace` and `Peeker` objects to create a new `Trace` as a result:

    <Trace|Peeker> op <Trace|Peeker|integer|float> => Trace

There is also a `delay()` method for creating a time-shifted version of a `Trace` (useful for
detecting edges):

    <Trace|Peeker>.delay(<integer>) => Trace

The times at which the resulting trace is `True` (i.e., non-zero) can be extracted as a list:

    <Trace|Peeker>.trig_times() => [t0, t1, ...]

These time values can be used to set the `start_time` parameter when displaying waveforms or tables:

```py
trigs = trc.trig_times()
Peeker.show_waveforms(..., start_time=trigs[0])
```

I'll demonstrate the creation of triggers using `Trace` objects containing random values.
(These same operations can be performed on `Peeker` objects;
it's just easier to create a `Trace`.)

````python
from random import randrange
from myhdlpeek import Trace, Sample, show_traces

def create_random_trace(name, num_bits, num_samples):
    trace = Trace()
    trace.name = name
    trace.num_bits = num_bits
    for i in range(num_samples):
        trace.append(Sample(i, randrange(0,2**num_bits)))
    return trace

trc1 = create_random_trace('Trc1', 4, 100)
trc2 = create_random_trace('Trc2', 1, 100)

Peeker.show_traces(trc1, trc2, stop_time=20);
````

A common trigger is to look for a positive-going edge on a signal.
This can be done by logically-ANDing the signal to a time-delayed and inverted version of itself:

````python
posedge_trc = trc2 & ~trc2.delay(1)    # Create trigger trace that is 1 whenever trc2 has a rising edge.
posedge_trc.name = 'Positive Edge'
trigs = posedge_trc.trig_times()       # Get times at which the trigger trace is 1.
print('Trigger times:', trigs)
start_time = trigs[0]                  # Start waveform display at the first trigger.
stop_time = start_time+20              # Stop waveform display after 20 time units.
Peeker.show_traces(trc2, posedge_trc, start_time=start_time, stop_time=stop_time, tock=True);
````

Naturally, there are convenience functions for these common cases: `posedge()`, `negedge()` and `anyedge()`:

````python
posedge_trc = trc2.posedge(); posedge_trc.name = 'Positive Edge'
negedge_trc = trc2.negedge(); negedge_trc.name = 'Negative Edge'
anyedge_trc = trc2.anyedge(); anyedge_trc.name = 'Any Edge'
Peeker.show_traces(trc2, posedge_trc, negedge_trc, anyedge_trc, stop_time=20);
````

Another common trigger is to look for when a bus has a certain value:

````python
val_trc = trc1 == 7  # Create a trigger trace that is 1 whenever trc1 has the value 7.
trigs = val_trc.trig_times()
Peeker.show_traces(trc1, val_trc, start_time=trigs[0]-5, stop_time=trigs[0]+15);
````

Or detect when a bus is between two values:

````python
val_trc = (5 <= trc1) & (trc1 <= 8)  # Trigger when trace value is in range [5,8].
trigs = val_trc.trig_times()
Peeker.show_traces(trc1, val_trc, start_time=trigs[0], stop_time=trigs[0]+20);
````

Or trigger when several bits of a bus have a certain value:

````python
bit_val_trc = ((trc1 & 0b0110)>>1) == 3 # Trigger when bits 1 and 2 are both on (trc1 is 6, 7, 14 or 15).
trig = bit_val_trc.trig_times()[0]
Peeker.show_traces(trc1, bit_val_trc, start_time=trig-5, stop_time=trig+15);
````

Complicated triggers are possible. Here's one that triggers when consecutive bus values differ by more than 10:

````python
diff_trc = (abs(trc1 - trc1.delay(1))) > 10
trig = diff_trc.trig_times()[0]
Peeker.show_traces(trc1, diff_trc, start_time=trig-5, stop_time=trig+15);
````

Since the output of operations on `Trace`s is another `Trace`, it's possible to
concatenate several operations to get a concise trigger expression:

````python
trig = ((5 <= trc1) & (trc1 <= 8)).trig_times()[0]
Peeker.show_traces(trc1, val_trc, start_time=trig, stop_time=trig+20);
````

## Using myhdlpeek in a Python Script

In addition to Jupyter notebooks, you can also use `myhdlpeek` in stand-alone Python scripts as shown below with the multiplexer example:

```py
from myhdl import * 
from myhdlpeek.myhdl import Peeker

# Define the multiplexer module.
def mux(z, a, b, sel):

    @always_comb
    def mux_logic():
        if sel == 1:
            z.next = a
        else:
            z.next = b
           
    return mux_logic

# Create some signals to attach to the multiplexer.
a, b, z = [Signal(0) for _ in range(3)]
sel = Signal(bool(0))

# Create some Peekers to monitor the multiplexer I/Os.
Peeker(a, 'a')
Peeker(b, 'b')
Peeker(z, 'z')
Peeker(sel, 'select')

# Instantiate the multiplexer.
mux_1 = mux(z, a, b, sel)

# Create a test process to apply random patterns to the multiplexer.
from random import randrange
def test():
    for _ in range(8):
        a.next, b.next, sel.next = randrange(8), randrange(8), randrange(2)
        yield delay(1)

# Create & run a simulator for the multiplexer, test process and the peekers.
Simulation(mux_1, test(), *Peeker.instances()).run()

# Display the complete waveforms captured by all the Peekers.
Peeker.show_waveforms()

# Show the waveforms in a dialog window.
import matplotlib.pyplot as plt
plt.show()

```
